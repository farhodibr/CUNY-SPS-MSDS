```python
np.random.seed(123)
```

```python

item_ratings = np.random.randint(1, 5, size=(10, 10)).astype(float)
```

```python
missing_ratings = np.random.rand(10, 10) < 0.3
item_ratings[missing_ratings] = np.nan
```

```python
user_ids = [f"User_{i + 1}" for i in range(10)]
item_ids = [f"Item_{x + 1}" for x in range(10)]
```

```python
toy_df = pd.DataFrame(
    item_ratings, 
    index=user_ids, 
    columns=item_ids
    )
```

```python
print(toy_df)
```

```python
toy_df.style\
    .set_caption("Toy Ratings Matrix (10 Users × 10 Items)")\
    .format(precision=2, na_rep="NaN")
    
```
